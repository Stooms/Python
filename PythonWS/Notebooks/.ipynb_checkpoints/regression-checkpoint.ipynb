```python
# On importe les librairies dont on aura besoin pour ce tp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d.axes3d import get_test_data
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
```

```python
# On charge le dataset
house_data = pd.read_csv('house_data.csv')

# Récupérer l'ensemble des valeurs de la variable cible
Y = house_data["price"]
# Récupérer les variables prédictives (on en a 2)
X = house_data[['surface','arrondissement']]

# On affiche le nuage de points dont on dispose
fig = plt.figure()
ax = fig.add_subplot(1,2,1, projection='3d')
ax.scatter(house_data["surface"], house_data["arrondissement"], house_data["price"], c='r', marker='^')
 
ax.set_xlabel('surface')
ax.set_ylabel('arrondissement')
ax.set_zlabel('prix')
 
plt.show()

#plt.plot(house_data['surface'], house_data['price'], 'ro', markersize=4)
#plt.plot(house_data['arrondissement'], house_data['price'], 'ro', markersize=4)
plt.gcf().set_size_inches(15,15)
#plt.show()
```
